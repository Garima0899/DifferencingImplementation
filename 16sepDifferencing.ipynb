A stationary time series is one whose statistucal properties  do not depend on the time at which the series is observed .Thus,time series series with trends,or with seasonality,are not stationary-trend and seasonality will affect the value of the series at different timess.
Stationary signifies that the statistical  properties of time series,such as mean,variance and covariance remain constant

Differencing: Take the difference between consecutive observation to remove the trend. This can be once or multiple time times until the data becomes stationary

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\hp\Downloads\AirPassengers.csv")
```

```python
df
```

```python
plt.rcParams.update({'figure.figsize':(12,6)})
df['#Passengers'].plot()
```

```python
df['#Passengers_diff'] = df['#Passengers']-df['#Passengers'].shift(1)
```

```python
df['#Passengers_diff'].dropna().plot()
```

```python
df['#Passengers_diff_7'] = df['#Passengers']-df['#Passengers'].shift(7)
```

```python
df['#Passengers_diff_7'].dropna().plot()
```

Method 2 TRANSFORMATION

```python
import numpy as np
df['adj_log']= np.log(df['#Passengers'])
df['adj_sqrt'] = np.sqrt(df['#Passengers'])
df['adj_cbrt'] = np.cbrt(df['#Passengers'])
```

```python
df['adj_log'].dropna().plot()
```

```python
df['adj_sqrt'].dropna().plot()
```

```python
df['adj_cbrt'].dropna().plot()
```

```python
df['#Passengers_log_diff'] = df['adj_log']-df['adj_log'].shift(1)
```

```python
df['#Passengers_log_diff'].dropna().plot()
```

```python
df['#Passengers_sqrt_diff'] = df['adj_sqrt']-df['adj_sqrt'].shift(1)
df['#Passengers_sqrt_diff'].dropna().plot()
```
